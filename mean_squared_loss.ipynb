```python
import numpy as np
```

### Forward Pass
$$
\mathrm{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_{\text{pred}, i} - y_{\text{true}, i})^2
$$


### Backward Propagation
$$
\frac{\partial \mathrm{MSE}}{\partial y_{\text{pred}, i}}
= \frac{2}{n} (y_{\text{pred}, i} - y_{\text{true}, i})
$$


```python
class MSE_loss:
    def __init__(self):
        self.y_pred=None
        self.y_true=None


    #forward pass

    def forward(self,y_pred,y_true):
        """
        y_pred => predication
        y_true => ground truth
        """

        self.y_pred=y_pred
        self.y_true=y_true

        n=y_true.shape[0]

        loss = np.sum((y_pred-y_true)**2)/n

        return loss
    
    def backward(self):

        """
        We now have to return the gradient w.r.t y_pred
        """

        n=self.y_true.shape[0]

        grad = (2/n)*(self.y_pred-self.y_true)

        return grad

```

```python
#Test the function
y_true = np.array([1.0, 2.0, 3.0])
y_pred = np.array([2.0, 2.5, 2.0])

loss_fn=MSE_loss()

loss=loss_fn.forward(y_pred,y_true)
grad=loss_fn.backward()

print("Loss:",loss)
print("Gradient:",grad)
```
